```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
```

```python
def show_img(img, title):
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot()
    imgplot = plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    ax.set_title(title)
    plt.show()
```

```python
low_detail = image.imread("low_detail.jpeg")
high_detail = image.imread("high_detail.jpeg")
low_contrast = image.imread("low_contrast.jpeg")
high_contrast = image.imread("high_contrast.jpeg")
```

```python
show_img(low_detail, 'Original low detail image')
show_img(high_detail, 'Original high detail image')
show_img(low_contrast, 'Original low contrast image')
show_img(high_contrast, 'Original high contrast image')
```

```python
def slide_hist_right(img, value):
    res_img = np.asarray(img, dtype=np.uint16)
    res_img = np.minimum(255, res_img + value)
    return np.asarray(res_img, dtype=np.uint8)
```

```python
show_img(low_detail, 'Before')
show_img(slide_hist_right(low_detail, 50), 'After')
```

```python
show_img(high_detail, 'Before')
show_img(slide_hist_right(high_detail, 50), 'After')
```

```python
show_img(low_contrast, 'Before')
show_img(slide_hist_right(low_contrast, 50), 'After')
```

```python
show_img(high_contrast, 'Before')
show_img(slide_hist_right(high_contrast, 50), 'After')
```
